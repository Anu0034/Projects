```python
df1=pd.get_dummies(df[['class','cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing',
'gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring',
'stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population']],drop_first=True)
df1
```

```python
df1.dtypes
```

```python
x=df1.iloc[:,:-1]
x
```

```python
y=df1.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
mat=confusion_matrix(y_test,y_pred)
print(classification_report(y_test,y_pred))
mat
```

```python
score=accuracy_score(y_test,y_pred)
score
```
