```python
df1=pd.get_dummies(df[['Sex','BP','Cholesterol']],drop_first=True)
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

```python
df2.drop(columns=['Sex','BP','Cholesterol'],inplace=True)
```

```python
df2
```

```python
df2.dtypes
```

```python
x=df2.drop(['Drug'],axis=1).values
x
```

```python
y=df2['Drug'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier =RandomForestClassifier(n_estimators=7,criterion='entropy',random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=7,criterion='gini',random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
