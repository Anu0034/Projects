```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[150,55,72]]))
```

```python
print("slope is")
list(zip(x,model.coef_))
```

```python
print("constant is",model.intercept_)
```

```python
df1=pd.DataFrame({'actual_value':y_test,'predicted_value':y_pred,'difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error
print('error is',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print("squared error is",mean_squared_error(y_test,y_pred))
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
print("r2 score is",r2_score(y_test,y_pred))
```
