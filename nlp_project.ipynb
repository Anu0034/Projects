```python
#positive : 1  negative : -1 neutral : 0
df['target']=df['target'].map({'Positive':1,'Neutral':0,'Negative':-1})
df
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
tweets=df.text
tweets
```

```python
#tokenization
from nltk import TweetTokenizer
tw=TweetTokenizer()
tweets=tweets.apply(lambda x:tw.tokenize(x)).apply(lambda x:" ".join(x)) #fun apply x-argument .join space to join
```

```python
tweets
```

```python
tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')# + combination of special characters remove
tweets
```

```python
#3>=words  do ws remove
from nltk.tokenize import word_tokenize
tweets=tweets.apply(lambda x:" ".join([w for w in word_tokenize(x) if len(w)>=3]))
tweets
```

```python
#stemming
from nltk.stem import SnowballStemmer
sb=SnowballStemmer('english')
tweets=tweets.apply(lambda x:[sb.stem(i.lower())for i in tw.tokenize(x)]).apply(lambda x:' '.join(x))
tweets
```

```python
#stopwords
from nltk.corpus import stopwords
sw=stopwords.words('english')
tweets=tweets.apply(lambda x:[i for i in tw.tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
tweets
```

```python
#vecotorization
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data=vec.fit_transform(tweets)
```

```python
train_data.shape
```

```python
print(train_data)
```

```python
y=df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
mat=confusion_matrix(y_test,y_pred)
score1=accuracy_score(y_test,y_pred)
mat
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
y_pred
```
